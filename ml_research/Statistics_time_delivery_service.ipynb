```python
# Функция считающая процентиль для агрегации
def percentile(n):
    def percentile_(sroki):
        return np.percentile(sroki, n)
    percentile_.__name__ = 'percentile_%s' % n
    return percentile_
```

```python
# Из типа объект делаем даты
sroki['ДатаПоступленияНаСклад'] = pd.to_datetime(sroki.ДатаПоступленияНаСклад, format='%d.%m.%Y %H:%M:%S')
sroki['ДокументРезерваДатаСоздания'] = pd.to_datetime(sroki.ДокументРезерваДатаСоздания, format='%d.%m.%Y %H:%M:%S')
```

```python
# Информация по датасету сроки доставки
sroki.info()
```

```python
# Отсортируем предиктор дата_поступления_на_склад
sroki = sroki.sort_values(by=['ДатаПоступленияНаСклад'])
```

```python
# Посмотрим на датасет
sroki
```

```python
# Выберем данные за последние 3 месяца
sroki = sroki[(sroki.ДатаПоступленияНаСклад.dt.month >= 2) & (sroki.ДатаПоступленияНаСклад.dt.month <= 4) & (sroki.ДатаПоступленияНаСклад.dt.year == 2020)]
sroki
```

```python
# Создадим предиктор со сроками поставок
sroki['difference'] = sroki.ДатаПоступленияНаСклад - sroki.ДокументРезерваДатаСоздания

# Посмотрим результат
sroki.head(10)
```

```python
# Выберем только количество 
# дней в сроках поставок
lst = []
for item in sroki.difference:
    lst.append(item.days)
sroki['delivery'] = lst

# Посмотрим информацию по датасету
sroki.info()
```

```python
# Результат предыдущих операций
sroki.head()
```

```python
# Сгруппируем датасет по производителям и складам, агрегирую сроки поставки процентилем
sroki_50 = sroki.groupby(['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',\
                         'Склад', 'СкладКод']).agg({'delivery': percentile(50)}).reset_index()
sroki_90 = sroki.groupby(['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',\
                         'Склад', 'СкладКод']).agg({'delivery': percentile(90)}).reset_index()
sroki_mean = sroki.groupby(['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',\
                            'Склад', 'СкладКод']).agg({'delivery': np.mean}).reset_index()
sroki_min = sroki.groupby(['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',\
                            'Склад', 'СкладКод']).agg({'delivery': np.min}).reset_index()
sroki_max = sroki.groupby(['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',\
                            'Склад', 'СкладКод']).agg({'delivery': np.max}).reset_index()
```

```python
sroki_mean['mean'] = sroki_mean.delivery
sroki_50['percentile_50'] = sroki_50.delivery
sroki_90['percentile_90'] = sroki_90.delivery
sroki_max['max'] = sroki_max.delivery
sroki_min['min'] = sroki_min.delivery

sroki_mean.drop('delivery', axis=1, inplace=True)
sroki_50.drop('delivery', axis=1, inplace=True)
sroki_90.drop('delivery', axis=1, inplace=True)
sroki_max.drop('delivery', axis=1, inplace=True)
sroki_min.drop('delivery', axis=1, inplace=True)
```

```python
perc50_perc90 = pd.merge(sroki_50, sroki_90,\
                 on = ['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод', 'Склад', 'СкладКод'], how = 'left')
max_min = pd.merge(sroki_max, sroki_min,\
                 on = ['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод', 'Склад', 'СкладКод'], how = 'left')
statistics = pd.merge(perc50_perc90, max_min, \
                on = ['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',  'Склад', 'СкладКод'], how = 'left')
statistics = pd.merge(statistics, sroki_mean, \
                on = ['Номенклатура_Производитель', 'Номенклатура_ПроизводительКод',  'Склад', 'СкладКод'], how = 'left')
statistics
```

```python
statistics.to_json('statistics.json', orient="split")
statistics.to_excel('statistics.xlsx', index=False)  
```
