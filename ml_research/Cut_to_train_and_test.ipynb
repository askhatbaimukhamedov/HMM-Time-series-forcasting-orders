```python
import os
import time
import datetime
import warnings
import numpy as np
import pandas as pd
```

```python
def add_date_block_num(data_frame):
    """
    Добавляет предикторы date_block_num и real_data
        :param data_frame: Таргетный датасет
        :return: None
    """
    lst_features = list()
    data_frame['real_date'] = pd.to_datetime(data_frame.date,
                                                 format='%d.%m.%Y %H:%M:%S')
    for item in data_frame.real_date:
        if item.year == 2018:
            lst_features.append(item.isocalendar()[1])
        if item.year == 2019:
            lst_features.append(item.isocalendar()[1] + 52)
        if item.year == 2020:
            lst_features.append(item.isocalendar()[1] + 104)

    data_frame['date_block_num'] = pd.Series(lst_features).astype('int')
```

```python
# Директория с датасетами
print(os.listdir('../data'))
```

```python
orders = pd.read_csv('../data/orders.csv', sep = ';')
```

```python
# Добавим предикторы (date_block_num, ral_date)
add_date_block_num(orders)
```

```python
# Уберем не полную неделю
orders = orders[orders['date_block_num'] <= 143]
orders
```

```python
# Разделим датасет на обучающий и тестовый наборы.
# В тесте данные за последнюю неделю
train = orders[orders['date_block_num'] <= 142]
test = orders[orders['date_block_num'] > 142]
```

```python
# Посмотрим на обучающий набор
train
```

```python
# Посмотрим на тестовый набор
test
```

```python
# Дропним добавленные предикторы
train.drop(['real_date', 'date_block_num'], axis=1, inplace=True)
test.drop(['real_date', 'date_block_num'], axis=1, inplace=True)
```

```python
# Сохраним полученные датасеты
train.to_csv('train_orders.csv', index = False)
test.to_csv('test_orders.csv', index = False)
```
